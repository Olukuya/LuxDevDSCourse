```python
import matplotlib.pyplot as plt     #when inputing the magic command it will show the visualization is displayed in this exact notebook. never put the comment on the magic command
%matplotlib inline    
```

```python
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [9, 8, 7, 6, 5, 4, 3, 2, 1]
```

line plot

should have title, labels on the different axes

```python
#line plot in matplotlib library
plt.plot(x,y)  #for a line plot it should be plt.plot. Just write plot
plt.xlabel('x - values') #naming x axis
plt.ylabel('y - values') #naming y axis
plt.title('Title') #naming of the title
plt.show() # shows visualization here.
```

```python
sns.lineplot(x=x, y=y)   #in seaborn library
plt.xlabel('x - values') #naming x axis
plt.ylabel('y - values') #naming y axis
plt.title('Title') #naming of the title
plt.show() # shows visualization here.
#plt can be used here in the naming
```

pie chart

```python
# pie chart show percentage of a whole 
```

```python
gender = ['Female' , 'Male']
counts = [50 , 50]
```

```python
#matplotlib   for pie chart it takes two main arguments: different numbers called sizes and labels
#research on the reason why sizes = counts does not work and why removing lables fails
plt.pie( counts, labels = gender, autopct = '%1.1f%%') #includes percentage, 1.1 shows .5 percent. If using 1.2 it will be .05 percent. autopct is autopercentage
plt.title('Gender Distribution') #naming
plt.show()
```

```python
#seaborn
#the equivalent of a pie chart in seaborn is called a pyplot
#research on it
#code will be sent for this particular one
#not giving the needed result
sns.countplot(x = counts, hue = gender)   #seaborn should have count, hue and the data 
plt.title('Gender Distribution', color = 'red')
plt.show()
```

Histogram

```python
#used to show only one variable 
#matplotlib
#bins divided the numbers to equal parts
#play with the number of bins to find out
plt.hist(y, bins = 5) #bins are the number of different collums
```

```python
#seaborn
#bins divided the numbers to equal parts
#play with the number of bins to find out: divides values into equal different parts in relation to the bins.
sns.histplot(x = y, bins = 9) #you can have only x or y or x = x or x = y
```
